```python
import pandas as pd
```

```python
df_gf = pd.read_csv('../dataset/train_fold.csv')
```

```python
df_gf
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=25)
```

```python
df_80 = df_gf[df_gf['fold'] != 0]
```

```python
for fold_number, (train_index, val_index) in enumerate(kfold.split(X=df_80.index, y=df_80['label_group'])):
    df_80.loc[df_80.iloc[val_index].index, 'fold_strat'] = fold_number
```

```python
df_80
```

```python
df_gf.loc[df_80.index, 'fold_strat'] = df_80['fold_strat']
```

```python
df_gf
```

```python
df_gf.rename(columns={'fold': 'fold_group'}, inplace=True)
```

```python
df_gf
```

```python
df_gf.to_csv('../dataset/reliable_validation_tm.csv', index=False)
```
